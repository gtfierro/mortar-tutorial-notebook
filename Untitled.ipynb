```python
import pymortar
import pandas as pd
client = pymortar.Client()
query = """
SELECT ?name ?type
WHERE {
    ?name rdf:type ?type
};
"""
qresp = client.qualify([query])
resp = client.fetch(pymortar.FetchRequest(
    sites=qresp.sites,
    views=[
        pymortar.View(
            name="t",
            definition=query,
        )
    ],
    time=pymortar.TimeParams(
        start="2018-01-01T00:00:00Z",
        end="2018-01-02T00:00:00Z",
    )
))
df = pd.DataFrame(resp.query('select * from t'))
print(df)
```
